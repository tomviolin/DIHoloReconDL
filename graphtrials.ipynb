```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os,sys
import time
print(os.getcwd())
```

```python
from glob import glob
from IPython.display import clear_output
def dochart():
    plt.close('all')
    dlist = sorted(glob('trials/trial*/training_log.csv'))
    df = pd.read_csv(dlist[-1])[:]
    mkrsize = np.float32(df['is_best'])*80+1
    plt.figure(figsize=(8,3))
    
    plt.plot(df.epoch,df['train_loss'],'.-',label="train")
    plt.plot(df.epoch,df['val_loss'],'.-',label="val")
    plt.scatter(x=df.epoch,y=df['val_loss'],s=mkrsize)
    plt.yscale('log')
    plt.legend()
    plt.show()
    xx=df['epoch']
    plt.figure(figsize=(5,2))
    plt.plot(df.epoch,df['learning_rate'],label='lr')
    plt.plot(xx,0.01*np.exp(-xx/70),lw=0.4,ls='--',label="lr bound 1")
    plt.plot(xx,0.05*np.exp(-xx/70),lw=0.4,ls='--',label="lr bound 2")
    plt.legend()
    plt.yscale('log')
    plt.show()

while True:
    clear_output(wait=True)
    dochart()
    time.sleep(5)
```

```python
#power
power = 2
# x range
x=np.arange(0,1,0.01);
# plot x to a power
plt.plot(x,x**power);

#plot 0.5, 0.5 to power
plt.plot(0.5,0.5**power,'.',ms=20);
# plot dotted lines from axes to point
plt.hlines(0.5**2,0,0.5,ls='--'); plt.vlines(0.5,0,0.5**2,ls='--')

#plot 0.5 ** (1/power), 0.5
plt.plot(0.5**(1/power),0.5,'.',ms=20);
# plot dotted lines from axes to point
plt.hlines(0.5,0,0.5**(1/power),ls='--'); plt.vlines(0.5**(1/power),0,0.5,ls='--')
```
